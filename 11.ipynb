```python
                                                            # Experiment 11

# (a) Employee Class
# ------------------------------------------

# Objective:
# To store employee details and count employees.

# Code:
class Employee:
    count = 0

    def __init__(self, name, designation, salary):
        self.name = name
        self.designation = designation
        self.salary = salary
        Employee.count += 1

e1 = Employee("Aman", "Manager", 50000)
e2 = Employee("Riya", "Developer", 40000)

print("Total Employees:", Employee.count)

# Code Explanation:
# Class variable count tracks total employees.
# Constructor initializes employee details.

# Result:
# Total number of employees is printed.


```

```python
# ------------------------------------------
# (b) Student, Course, Department Classes
# ------------------------------------------

# Objective:
# To enroll a student in a course and department.

# Code:
class Department:
    def __init__(self, name):
        self.name = name

class Course:
    def __init__(self, name, year):
        self.name = name
        self.year = year

class Student:
    def __init__(self, name, rollno, course, year):
        self.name = name
        self.rollno = rollno
        self.course = course
        self.year = year

dept = Department("CSE")
course = Course("AI", 2025)
student = Student("Anjali", 101, course.name, course.year)

print("Student enrolled successfully")

# Code Explanation:
# Separate classes represent Department, Course and Student.
# Student is linked with course and department.

# Result:
# Enrollment message is displayed.


```

```python
# ------------------------------------------
# (c) Exception Handling for Integer Input
# ------------------------------------------

# Objective:
# To handle invalid integer input using exception handling.

# Code:
while True:
    try:
        num = int(input("Enter an integer: "))
        print("Valid integer entered:", num)
        break
    except ValueError:
        print("Invalid input, please enter an integer.")

# Code Explanation:
# try block checks for valid integer input.
# except block handles invalid input error.

# Result:
# Program continues until valid integer is entered.

```

```python
                                                        # Experiment:5
# (a) Number Pattern using Nested Loops

# Objective:
# To print number pattern using nested loops.

# Code:
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end="")
    print()

# Result:
# Pattern is printed as required.
```

```python
# ------------------------------------------
# (b) Multiplication Table (Again)
# ------------------------------------------

# Objective:
# To print multiplication table using loop.

# Code:
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{n} x {i} = {n*i}")

# Result:
# Table is displayed.
```

```python
# ------------------------------------------
# (c) Keyword Arguments Function
# ------------------------------------------

# Objective:
# To demonstrate keyword arguments in function.

# Code:
def student_info(name, age, course):
    print("Name:", name)
    print("Age:", age)
    print("Course:", course)

# Function call using keyword arguments
student_info(course="AI", name="Anjali", age=22)

# Result:
# Student information is printed.
```

```python
# ------------------------------------------
# (d) Absolute, Square Root and Cube
# ------------------------------------------

# Objective:
# To calculate absolute value, square root and cube.

# Code:
import math

num = float(input("Enter a number: "))

print("Absolute Value:", abs(num))
print("Square Root:", math.sqrt(abs(num)))
print("Cube:", num ** 3)

# Result:
# Absolute value, square root and cube are displayed.
```

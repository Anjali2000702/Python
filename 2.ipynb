```python

                                                #Experiment 2

#(a) Area of Triangle using Heron’s Formula

#Objective:
#To calculate the area of a triangle using Heron’s Formula based on the lengths of its sides.
#Code:

import math

#Input sides of the triangle

a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

#Calculate semi-perimeter

s = (a + b + c) / 2

#Apply Heron's Formula

area = math.sqrt(s * (s - a) * (s - b) * (s - c))

#Display result

print("Area of the triangle:", area)

#Code Explanation:
#The user enters the three sides of the triangle.
#The semi-perimeter is calculated as (a + b + c) / 2.
#Heron’s formula is applied to compute the area.
#The math.sqrt() function is used to find the square root.
#Result:
#The program displays the area of the triangle based on the given sides.


```

```python
#(b) Total Amount of Money in Piggy Bank

#Objective:
#To calculate the total amount of money in a piggy bank containing coins of Rs 10, 5, 2, and 1.
#Code:
#Number of coins

rs10 = int(input("Enter number of Rs 10 coins: "))
rs5 = int(input("Enter number of Rs 5 coins: "))
rs2 = int(input("Enter number of Rs 2 coins: "))
rs1 = int(input("Enter number of Rs 1 coins: "))

#Calculate total amount

total_amount = (rs10 * 10) + (rs5 * 5) + (rs2 * 2) + (rs1 * 1)

#Display result

print("Total amount in piggy bank: Rs", total_amount)

#Code Explanation:
#The user inputs the number of coins for each denomination.
#Each coin count is multiplied by its respective value.
#The total sum is calculated using arithmetic operators.
#Result:
#The program outputs the total money present in the piggy bank.
```

```python
                                                    # Experiment 12

# (a) Operator Overloading
# ------------------------------------------

# Objective:
# To overload + operator for Student class.

# Code:
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

s1 = Student("A", 80)
s2 = Student("B", 70)

print("Total Marks:", s1 + s2)

# Code Explanation:
# __add__ method overloads + operator.
# Adds marks of two Student objects.

# Result:
# Sum of marks is printed.


```

```python
# ------------------------------------------
# (b) Voting Eligibility Validation
# ------------------------------------------

# Objective:
# To check whether a person can vote.

# Code:
name = input("Enter name: ")
age = int(input("Enter age: "))

if age >= 18:
    print(name, "can vote")
else:
    print(name, "cannot vote")

# Code Explanation:
# Age is checked using conditional statement.

# Result:
# Voting eligibility is displayed.


```

```python
# ------------------------------------------
# (c) Exception Handling for Integer Input
# ------------------------------------------

# Objective:
# To handle invalid integer input using exception handling.

# Code:
while True:
    try:
        num = int(input("Enter an integer: "))
        print("Valid integer entered:", num)
        break
    except ValueError:
        print("Invalid input, please enter an integer.")

# Code Explanation:
# try block checks for valid integer input.
# except block handles invalid input error.

# Result:
# Program continues until valid integer is entered.


```

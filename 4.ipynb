```python
                                                        #Experiment 4
# (a) Operator Precedence Expression
# ------------------------------------------

# Objective:
# To evaluate an expression using operator precedence rules.

# Code:
# Given expression
x = 10 + 2 * 3 ** 2 - 8 / 4

# Step-wise calculation
step1 = 3 ** 2
step2 = 2 * step1
step3 = 8 / 4
step4 = 10 + step2
x = step4 - step3

# Display result
print("Final value of x:", x)

# Code Explanation:
# Exponentiation is evaluated first, then multiplication,
# followed by division, addition and subtraction.

# Result:
# The program prints the final value of x.


```

```python
# ------------------------------------------
# (b) Dictionary of Students and Marks
# ------------------------------------------

# Objective:
# To store student names and marks using dictionary and retrieve marks.

# Code:
students = {
    "Aman": 85,
    "Riya": 90,
    "Neha": 78
}

# Print dictionary
print("Student Dictionary:", students)

# Get marks of a student
name = input("Enter student name: ")

if name in students:
    print("Marks:", students[name])
else:
    print("Student not found")

# Code Explanation:
# Dictionary stores key-value pairs.
# User input is matched with dictionary keys.

# Result:
# Marks are printed if student exists.


```

```python
# ------------------------------------------
# (c) Reverse of a Number
# ------------------------------------------

# Objective:
# To print the reverse of a number.

# Code:
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    reverse = reverse * 10 + num % 10
    num //= 10

print("Reversed number:", reverse)

# Result:
# The reversed number is displayed.
```

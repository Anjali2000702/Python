```python
                                                        # Experiment 9
# (a) Count Spaces, Tabs and Newlines
# ------------------------------------------

# Objective:
# To count spaces, tabs and newlines in a file.

# Code:
with open("1.ipynb", "r") as f:
    data = f.read()

print("Spaces:", data.count(" "))
print("Tabs:", data.count("\t"))
print("Newlines:", data.count("\n"))

# Code Explanation:
# File is opened in read mode.
# count() function counts spaces, tabs and newlines.

# Result:
# Counts are displayed.


```

```python
# ------------------------------------------
# (b) Copy First 10 Bytes of Binary File
# ------------------------------------------

# Objective:
# To copy first 10 bytes of a binary file.

# Code:
with open("2.ipynb", "rb") as src:
    data = src.read(10)

with open("3.ipynb", "wb") as tgt:
    tgt.write(data)

print("First 10 bytes copied.")

# Code Explanation:
# Binary file is opened in read-binary mode.
# First 10 bytes are read and written to another file.

# Result:
# Binary data is copied successfully.
```

```python
                                                            #Experiment :6

# (a) Lambda Function to Square a Number
# ------------------------------------------

# Objective:
# To use a lambda function to calculate the square of a number.

# Code:
num = int(input("Enter a number: "))

square = lambda x: x * x

print("Square of the number:", square(num))

# Code Explanation:
# The input number is taken from the user.
# A lambda function is defined to compute square.
# The lambda function is called and result is printed.

# Result:
# The square of the given number is displayed.


```

```python
# ------------------------------------------
# (b) Recursive Function to Count Digits
# ------------------------------------------

# Objective:
# To count the number of digits in a positive integer using recursion.

# Code:
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)

num = int(input("Enter a positive integer: "))
print("Number of digits:", count_digits(num))

# Code Explanation:
# The function calls itself by reducing the number using integer division.
# Each recursive call counts one digit.
# Recursion stops when n becomes 0.

# Result:
# Total number of digits is printed.


```

```python
# ------------------------------------------
# (c) Reverse List without reverse() or slicing
# ------------------------------------------

# Objective:
# To reverse a list without using reverse() or slicing.

# Code:
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input("Enter element: ")))

print("Original List:", lst)

rev = []
for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i])

print("Reversed List:", rev)

# Code Explanation:
# Elements are taken into a list.
# A loop runs from last index to first index.
# Elements are appended to a new list in reverse order.

# Result:
# Original and reversed list are displayed.


```

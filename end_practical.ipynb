```python
#A dentist has different services with fixed prices
#The customer sees all services
#The customer chooses one service
#The program shows the total amount to pay
#If the customer pays in advance, they get 50% discount
#the program prints:
#Service name
#Original price
#Discount (if any)
#Final payable amount
#print("Logic of the Program")
#print("1️ Store all dentist services and prices")
#print("2️ Display services to the customer")
#print("3️ Ask the customer to choose a service")
#print("4️ Ask if payment is advance or not")
#print("5️ If advance payment → apply 50% discount")
#print("6️ Print the final bill")
# Step 1: Store dentist services and prices
services = {
    1: ("Root Canal Therapy", 250),
    2: ("Oral Hygiene Check", 50),
    3: ("Emergency Injury Treatment", 100),
    4: ("Post-Procedure Check-up", 150),
    5: ("Routine Check-ups and Consultation", 75)# dictionary create
}


print("Welcome to Dental Care Clinic")
print("Available Services In Our Clinic:")
# services provided
for key, value in services.items():
    print(key, value[0], "- $", value[1])

# Ask user to choose the service
choice = int(input("Enter the service number: "))

# Advance payment
pay_advance = input("Are you paying in advance? (yes/no): ")

#  Get selected service details
service_name, price = services[choice]

#  Apply discount if advance payment
if pay_advance.lower() == "yes":
    discount = price * 0.5
else:
    discount = 0

# Step 7: Calculate final amount
final_amount = price - discount

# Step 8: Display bill
print("\n----- BILL For The Selected Service-----")
print("Selected service by the user:", service_name)
print("Original Price of the service: -$", price)
print("Discount given to the user: $", discount)
print("Total Payable amount: $", final_amount)
```

```python
#b) Write a
#Python program to find the XOR of two given strings interpreted as binary
#numbers.

#              Input:

#               ['0001','1011']

#              Output:

#             0b1010

num1 = '0001' # number is in binary 
num2 = '1011' # '' is the string type
print(type(num1))
print("First Number is:", num1)
print("Second Number is:",num2)
result = int(num1, 2) ^ int(num2, 2)# convert number in int and perform XOR
#For XOR same bit:0 Different bit 1
#0b to convert int to binary
print(bin(result))
```

```python
#Given two points (x1, y1) and (x2,
#y2), return the slope and intercept of the line joining these two points (the
#line may be perfectly horizontal or vertical). Input should be in this format:
#(x1, y1, x2, y2).
def line_from_two_points(x1, y1, x2, y2):
    # Case 1: Vertical line
    if x1 == x2:
        slope = None  # ya float('inf')
        intercept = None  # y-axis se nahi kaategi
        return slope, intercept
    
    # Case 2: Normal line (vertical nahi hai)
    slope = (y2 - y1) / (x2 - x1)
    
    # Intercept nikalo: c = y - mx
    intercept = y1 - slope * x1
    
    return slope, intercept

# Example chala ke dekhte hain
if __name__ == "__main__":
    # Test 1: Normal line
    x1, y1, x2, y2 = 1, 2, 3, 4
    m, c = line_from_two_points(x1, y1, x2, y2)
    print(f"Points: ({x1},{y1}), ({x2},{y2})")
    print(f"Slope: {m}, Intercept: {c}")
    print(f"Line equation: y = {m}x + {c}")
    print()
    
    # Test 2: Horizontal line (y same)
    x1, y1, x2, y2 = 2, 5, 7, 5
    m, c = line_from_two_points(x1, y1, x2, y2)
    print(f"Points: ({x1},{y1}), ({x2},{y2})")
    print(f"Slope: {m}, Intercept: {c}")
    print(f"Line equation: y = {c}")
    print()
    
    # Test 3: Vertical line (x same)
    x1, y1, x2, y2 = 4, 1, 4, 9
    m, c = line_from_two_points(x1, y1, x2, y2)
    print(f"Points: ({x1},{y1}), ({x2},{y2})")
    print(f"Slope: {m}, Intercept: {c}")
    print(f"Line equation: x = {x1}")
```

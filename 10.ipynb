```python
                                                            #Experiment 10
# (a) Remove Duplicates from List
# ------------------------------------------

# Objective:
# To remove duplicate elements from a list.

# Code:
n = int(input("Enter number of elements: "))
lst = []

for i in range(n):
    lst.append(int(input("Enter element: ")))

new_list = list(set(lst))

print("Original List:", lst)
print("Without Duplicates:", new_list)

# Code Explanation:
# Elements are stored in a list.
# set() removes duplicates automatically.
# Converted back to list.

# Result:
# List without duplicates is displayed.


```

```python
# ------------------------------------------
# (b) Tuple Index Search
# ------------------------------------------

# Objective:
# To find index of a number in a tuple.

# Code:
nums = (10, 20, 30, 40, 50)
x = int(input("Enter number: "))

if x in nums:
    print("Index:", nums.index(x))
else:
    print("Number not present")

# Code Explanation:
# Tuple is searched using in operator.
# index() returns position if found.

# Result:
# Index or not present message is shown.


```

```python
# ------------------------------------------
# (c) Set Operations
# ------------------------------------------

# Objective:
# To find common and symmetric difference of two sequences.

# Code:
a = list(map(int, input("Enter sequence 1: ").split()))
b = list(map(int, input("Enter sequence 2: ").split()))

set1 = set(a)
set2 = set(b)

print("Common Elements:", sorted(set1 & set2))
print("Symmetric Difference:", sorted(set1 ^ set2))

# Code Explanation:
# Lists are converted to sets.
# Intersection and symmetric difference are computed.

# Result:
# Required sets are printed.

```

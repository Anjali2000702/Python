```python
                                                        #Experiment 8


# (a) Remove Vowels from String
# ------------------------------------------

# Objective:
# To remove all vowels from a given string.

# Code:
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
result = ""

for ch in s:
    if ch not in vowels:
        result += ch

print("String without vowels:", result)

# Code Explanation:
# The string is scanned character by character.
# If character is not a vowel, it is added to result.

# Result:
# String without vowels is printed.


```

```python
# ------------------------------------------
# (b) Palindrome Check
# ------------------------------------------

# Objective:
# To check whether a string is palindrome ignoring case and symbols.

# Code:
import re

s = input("Enter a string: ")
clean = re.sub('[^a-zA-Z0-9]', '', s).lower()

if clean == clean[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# Code Explanation:
# Non-alphanumeric characters are removed.
# String is converted to lowercase.
# The cleaned string is compared with its reverse.

# Result:
# Palindrome or Not Palindrome is displayed.

```

```python
# ------------------------------------------
# (c) Reverse Words Order
# ------------------------------------------

# Objective:
# To reverse the order of words in a sentence.

# Code:
sentence = input("Enter a sentence: ")
words = sentence.split()

print("Reversed sentence:", " ".join(words[::-1]))

# Code Explanation:
# Sentence is split into words.
# Word order is reversed and joined back.

# Result:
# Sentence with reversed word order is printed.


```

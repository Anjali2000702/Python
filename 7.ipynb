```python
                                                        # Experiment:7


# (a) Random Module using Alias
# ------------------------------------------

# Objective:
# To generate random numbers using random module alias.

# Code:
import random as rnd

for i in range(5):
    print(rnd.randint(1, 100))

# Code Explanation:
# random module is imported as rnd.
# randint() generates random numbers between 1 and 100.
# Loop runs 5 times.

# Result:
# Five random integers are printed.

```

```python
# ------------------------------------------
# (b) Line Graph using matplotlib
# ------------------------------------------

# Objective:
# To plot a simple line graph using matplotlib.

# Code:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Graph")
plt.show()

# Code Explanation:
# pyplot is imported from matplotlib.
# x and y lists define data points.
# plot() draws the graph and show() displays it.

# Result:
# A line graph is displayed.


```

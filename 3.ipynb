```python

                                                # Experiment 3


# ------------------------------------------
# (a) Decimal to Binary, Octal and Hexadecimal
# ------------------------------------------

# Objective:
# To convert a decimal (base 10) integer into binary, octal, and hexadecimal
# using Python built-in functions.

# Code:
# Taking decimal number input from user
decimal_number = int(input("Enter an integer in decimal (base 10): "))

# Conversions using built-in functions
binary_value = bin(decimal_number)
octal_value = oct(decimal_number)
hexadecimal_value = hex(decimal_number)

# Display results
print("Binary equivalent:", binary_value)
print("Octal equivalent:", octal_value)
print("Hexadecimal equivalent:", hexadecimal_value)

# Code Explanation:
# The input() function takes a decimal number from the user.
# bin() converts decimal to binary.
# oct() converts decimal to octal.
# hex() converts decimal to hexadecimal.
# The converted values are printed.

# Result:
# The program displays the binary, octal, and hexadecimal equivalents
# of the given decimal number.

# GitHub Link:
# https://github.com/your-username/python-lab-experiment-3




```

```python
# ------------------------------------------
# (b) String to Float Conversion and Addition
# ------------------------------------------

# Objective:
# To read a string representing a floating-point number, convert it to float,
# add 10.5 to it, and display the result.

# Code:
# Taking string input representing a floating-point number
string_number = input("Enter a floating-point number as string: ")

# Convert string to float
float_number = float(string_number)

# Add 10.5
result = float_number + 10.5

# Display result
print("Result after adding 10.5:", result)

# Code Explanation:
# input() reads the number as a string.
# float() converts the string into a floating-point number.
# Arithmetic addition is performed using + operator.
# The final result is printed.

# Result:
# The program outputs the numeric result after adding 10.5
# to the converted floating-point number
```

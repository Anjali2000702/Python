```python
                                                    #Experiment 1

#(a) Display data of different types using variables and literal constants

#Objective

#To demonstrate the use of variables and literal constants of different data types in Python.

#Code 
# Different data types using variables
integer_var = 10
float_var = 25.75
string_var = "Machine Learning"
boolean_var = True
complex_var = 3 + 4j

# Displaying values and their data types
print("Integer Value:", integer_var, "Type:", type(integer_var))
print("Float Value:", float_var, "Type:", type(float_var))
print("String Value:", string_var, "Type:", type(string_var))
print("Boolean Value:", boolean_var, "Type:", type(boolean_var))
print("Complex Value:", complex_var, "Type:", type(complex_var))

# Literal constants
print("\nLiteral Constants:")
print(100)          # Integer literal
print(3.14)         # Float literal
print("Python")     # String literal
print(False)        # Boolean literal

#Code Explanation

#Variables are declared for different data types: integer, float, string, boolean, and complex.

#type() function is used to identify the data type of each variable.

#Literal constants are values written directly without assigning to a variable.

#Result

#The program displays values of different data types along with their corresponding Python data types.

```

```python
#(b) Application of the += operator on strings
#Objective

#To demonstrate the use of the += assignment operator on string variables.

#Code 
# Initial string
text = "Hello"

# Using += operator
text += " World"
text += "!"

print("Final String:", text)

#Code Explanation

#A string variable text is initialized.

#The += operator appends new strings to the existing string.

#Python supports string concatenation using +=.

#Result

#The output displays the concatenated string:

#Hello World!
```

```python
#(c) Addition and multiplication on string variables
#Objective

#To demonstrate addition (concatenation) and multiplication (repetition) operations on string variables.

#Code 
str1 = "AI"
str2 = "Lab"

# String addition (concatenation)
addition_result = str1 + " " + str2

# String multiplication (repetition)
multiplication_result = str1 * 3

print("String Addition Result:", addition_result)
print("String Multiplication Result:", multiplication_result)

#Code Explanation

#+ operator joins two strings (concatenation).

#* operator repeats the string specified number of times.

#These operations are supported only when operands are strings and integers.

#Result
#String Addition Result: AI Lab
#String Multiplication Result: AIAIAI 
```
